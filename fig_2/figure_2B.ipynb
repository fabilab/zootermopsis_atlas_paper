```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Close any open figures
plt.close('all')

# Reset seaborn + matplotlib to defaults
try:
    sns.reset_defaults()
except Exception:
    pass
mpl.rcdefaults()
plt.style.use('default')  # extra belt-and-braces
```

```python
palette = sns.color_palette("Set3", n_colors=4)
castes = ["queen", "king", "soldier", "worker"]
palette_dict = dict(zip(castes, palette))
```

```python
fig, axs = plt.subplots(1, 1, figsize=(9.5, 8))
axs.set_axis_off()
sc.pl.umap(
    adata,
    color='caste',
    ax=axs,
    add_outline=True,
    s=30,
    show=False, 
    title="",
    legend_loc=None,
    palette=palette_dict
    
)


fig.tight_layout()

plt.show()
```

```python
# Create a 2x2 subplot grid
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

# Flatten axes array for easy indexing
axes = axes.flatten()

# Loop through castes and plot each UMAP on a different subplot
for i, caste in enumerate(castes):
    sc.pl.umap(adata, color='caste', groups=[caste], ax=axes[i], show=False, 
               frameon=False, legend_loc=None, title='', add_outline=True, s=30,)

# Adjust layout to avoid overlap
plt.tight_layout()

# Display the figure
plt.show()
```
