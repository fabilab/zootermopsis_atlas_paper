```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Close any open figures
plt.close('all')

# Reset seaborn + matplotlib to defaults
try:
    sns.reset_defaults()
except Exception:
    pass
mpl.rcdefaults()
plt.style.use('default')  # extra belt-and-braces
```

```python
palette = sns.color_palette("Set3", n_colors=4)
castes = ["queen", "king", "soldier", "worker"]
palette_dict = dict(zip(castes, palette))
```

```python
## sensory neuron 

cell_cluster = adata[adata.obs['paper_cell_type_annotation'].isin(['sensory neuron'])]

fig, axes = plt.subplots(1, 4, figsize=(5.5, 1.5)) 
axes = axes.flatten()  

unique_castes = ["worker", "soldier", "king", "queen"]

for idx, caste in enumerate(unique_castes):
    with rc_context({"figure.figsize": (4, 4)}):
        sc.pl.umap(
            cell_cluster, 
            color='caste', 
            groups=[caste], 
            ax=axes[idx], 
            show=False, 
            legend_loc=None, 
            frameon=False,
            add_outline=True, 
            palette=palette_dict,
            size=100
        )
        axes[idx].set_title('')  

        
        axes[idx].set_xlim(-4.0, -0.9)
        axes[idx].set_ylim(8.1, 11.3)   



plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import scanpy as sc
import matplotlib.pyplot as plt

fat_marker = ['Znev00012632', 'Znev00004477']

cell_type = 'fat cell'
genes = fat_marker

adata_sub = adata[adata.obs['paper_cell_type_annotation'] == cell_type].copy()

# Get fly symbols if available
gene_labels = [
    adata.var.loc[g, 'dmel_gene_symbol']
    if pd.notna(adata.var.loc[g, 'dmel_gene_symbol']) else g
    for g in genes
]

# ---- Plot one UMAP per gene ----
for gene_id, gene_label in zip(genes, gene_labels):

    fig = sc.pl.umap(
        adata_sub,
        color=gene_id,
        frameon=False,
        add_outline=True,
        size=500,        # 🔥 change this to control point size
        vmin=0,
        vmax=4,
        return_fig=True,
        show=False,
        title=""
    )

    # Access colorbar (last axis)
    cbar = fig.axes[-1]
    cbar.set_yticks([0, 1, 2, 3, 4])   # 🔥 force clean ticks
    # ---- Make colourbar thicker (wider) ----
    pos = cbar.get_position()
    cbar.set_position([
        pos.x0,              # left
        pos.y0,              # bottom
        pos.width * 50,     # 🔥 increase thickness here
        pos.height           # keep height same
    ])

    # Optional: make colourbar outline + ticks thicker
    cbar.tick_params(width=1.5, length=4)
    for spine in cbar.spines.values():
        spine.set_linewidth(1.5)

    fig.set_size_inches(5.5, 5.0)

    fig.tight_layout()
    plt.show()
```
