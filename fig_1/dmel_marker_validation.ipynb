```python
import matplotlib.pyplot as plt

# Optional: Sort by logFC to pick most relevant genes
top_genes = de_with_orthologs.head(20)

# Number of UMAPs to show
genes_to_plot = top_genes['gene'].tolist()

# Layout for subplots
ncols = 4
nrows = -(-len(genes_to_plot) // ncols)  # Ceiling division

# Create figure
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols * 3, nrows * 3))
axes = axes.flatten()

# Plot each gene
for i, gene in enumerate(genes_to_plot):
    sc.pl.umap(
        adata,
        color=gene,
        ax=axes[i],
        show=False,
        frameon=False,
        title=gene,
    )
    axes[i].set_title(gene, fontsize=8)

# Turn off unused axes
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.suptitle("Top DE Genes in Muscle with Fly Orthologs", fontsize=14)
plt.subplots_adjust(top=0.92)
plt.show()
```
