```python
# -----------------------
# USER SETTINGS
# -----------------------
gene_info = {
    "Znev00002332": "PGRP – peptidoglycan-recognition protein sensing bacterial cell walls.",
    "Znev00007353": "GNBP – β-1,3-glucan binding PRR; detects fungal cell wall components.",
    "Znev00011578": "cSP (stubble) – serine protease involved in Toll/PO cascade activation.",
    "Znev00008234": "Easter – serine protease that processes Spaetzle to activate Toll signalling.",
    "Znev00006305": "Toll-like receptor – transmembrane receptor initiating Toll pathway signalling.",
    "Znev00006857": "spaetzle",
}

cell_type_col = "paper_cell_type_annotation"
caste_col = "caste"
castes = ["worker", "soldier", "king", "queen"]

MIN_FRAC = 0.10
MEAN_MIN = 0.01
EXPR_POS_THRESH = 0.01

palette = sns.color_palette("Set3", n_colors=4)
palette_dict = {
    "worker":  palette[3],
    "soldier": palette[2],
    "king":    palette[1],
    "queen":   palette[0],
}

# -----------------------
# Helpers for bar-chart task
# -----------------------
def _gene_vec(adata, gene_id):
    x = adata[:, gene_id].X
    if hasattr(x, "toarray"):
        x = x.toarray()
    return np.asarray(x).reshape(-1)

def count_celltypes_passing(adata, gene_id,
                           min_frac=0.10,
                           mean_min=0.50,
                           expr_pos_thresh=0.0,
                           cell_type_col="paper_cell_type_annotation",
                           caste_col="caste",
                           castes=("worker","soldier","king","queen")):
    expr = _gene_vec(adata, gene_id)
    df = adata.obs[[cell_type_col, caste_col]].copy()
    df["_expr"] = expr
    df["_pos"] = df["_expr"] > expr_pos_thresh

    out = {}
    for caste in castes:
        sub = df[df[caste_col] == caste]
        stats = (
            sub.groupby(cell_type_col, observed=False)
               .agg(mean_expr=("_expr", "mean"), frac_pos=("_pos", "mean"))
        )
        passing = (stats["frac_pos"] >= min_frac) & (stats["mean_expr"] >= mean_min)
        out[caste] = int(passing.sum())
    return out

def plot_celltype_counts_per_gene(
    adata,
    gene_ids,
    min_frac=0.10,
    mean_min=0.50,
    expr_pos_thresh=0.0,
    cell_type_col="paper_cell_type_annotation",
    caste_col="caste",
    castes=("worker","soldier","king","queen"),
    palette_dict=None,
    figsize=(7.0, 2.6),
    bar_edgecolor="k",
    bar_linewidth=0.9,
):
    rows = []
    for g in gene_ids:
        counts = count_celltypes_passing(
            adata, g,
            min_frac=min_frac,
            mean_min=mean_min,
            expr_pos_thresh=expr_pos_thresh,
            cell_type_col=cell_type_col,
            caste_col=caste_col,
            castes=castes
        )
        for caste in castes:
            rows.append({"gene": g, "caste": caste, "n_cell_types": counts[caste]})
    df_counts = pd.DataFrame(rows)

    genes = list(gene_ids)
    x = np.arange(len(genes))
    n_castes = len(castes)
    width = 0.18

    fig, ax = plt.subplots(figsize=figsize)

    # ✅ ensure grid is behind artists
    ax.set_axisbelow(True)

    for i, caste in enumerate(castes):
        vals = (
            df_counts[df_counts["caste"] == caste]
            .set_index("gene")
            .loc[genes, "n_cell_types"]
            .values
        )
        ax.bar(
            x + (i - (n_castes - 1) / 2) * width,
            vals,
            width=width,
            label=caste,
            color=(palette_dict[caste] if palette_dict else None),
            edgecolor=bar_edgecolor,
            linewidth=bar_linewidth,
            zorder=3,              # ✅ bars above grid
        )

    ax.set_xticks(x)
    ax.set_xticklabels(genes, rotation=45, ha="right")
    ax.set_xlabel("")
    ax.set_ylabel("# cell types passing")
    ax.set_title("", pad=8)

    ax.legend(frameon=False, ncol=4, bbox_to_anchor=(1.02, 1), loc="upper left")

    ax.grid(True, axis="y", alpha=0.3, zorder=0)  # ✅ grid behind bars

    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    fig.tight_layout()
    return fig, ax, df_counts

# -----------------------
# Run
# -----------------------
fig, ax, df_counts = plot_celltype_counts_per_gene(
    adata=adata,
    gene_ids=list(gene_info.keys()),
    min_frac=MIN_FRAC,
    mean_min=MEAN_MIN,
    expr_pos_thresh=EXPR_POS_THRESH,
    cell_type_col=cell_type_col,
    caste_col=caste_col,
    castes=castes,
    palette_dict=palette_dict,
    figsize=(7.2, 2.6),
    bar_edgecolor="k",
    bar_linewidth=0.9,
)
# 🔥 Save high quality (same style as before)
fig.savefig(
    "plots/figure_immune_celltype_counts.png",
    dpi=600,
    bbox_inches="tight"
)

plt.show()

df_counts
```

```python
## fig 4 E

import math
import matplotlib.pyplot as plt
import scanpy as sc

# -----------------------
# SETTINGS
# -----------------------
cell_types_to_show = ["hemocyte", "crystal cell"]
celltype_col = "paper_cell_type_annotation"

genes_to_plot = list(gene_info.keys())

N_COLS = 3
FIGSIZE_PER_PANEL = (2.4, 2.2)
POINT_SIZE = 35
ADD_OUTLINE = True

USE_FIXED_LIMITS = False

# -----------------------
# SUBSET cells
# -----------------------
adata_sub = adata[adata.obs[celltype_col].isin(cell_types_to_show)].copy()

# -----------------------
# Keep only genes present
# -----------------------
genes_present = [g for g in genes_to_plot if g in adata_sub.var_names]
if not genes_present:
    raise ValueError("None of the selected genes are present in adata.var_names.")

# -----------------------
# Short display names for titles
# -----------------------
def short_name(g):
    name = gene_display_names.get(g, g)
    return name.split("(")[0].strip()

# -----------------------
# Plot grid
# -----------------------
n = len(genes_present)
n_rows = math.ceil(n / N_COLS)

fig_w = FIGSIZE_PER_PANEL[0] * N_COLS
fig_h = FIGSIZE_PER_PANEL[1] * n_rows
fig, axes = plt.subplots(n_rows, N_COLS, figsize=(fig_w, fig_h))

if n_rows == 1:
    axes = [axes]

axes_flat = [ax for row in axes for ax in (row if hasattr(row, "__len__") else [row])]

for i, gene in enumerate(genes_present):
    ax = axes_flat[i]
    ax.set_axis_off()

    sc.pl.umap(
        adata,
        color=gene,
        ax=ax,
        s=POINT_SIZE,
        add_outline=ADD_OUTLINE,
        show=False,
        title=short_name(gene),
        vmax=4.5
    )

    XLIM = (0, 7)
    YLIM = (12.5, 16.5)

    if USE_FIXED_LIMITS:
        ax.set_xlim(*XLIM)
        ax.set_ylim(*YLIM)

# Turn off empty panels
for j in range(i + 1, len(axes_flat)):
    axes_flat[j].set_axis_off()

fig.tight_layout()

# 🔥 Save high resolution
fig.savefig(
    "plots/figure_4E_umap_grid.png",
    dpi=600,
    bbox_inches="tight"
)

plt.show()
```

```python
# supplementary 4H
# Figure 4H — Dotplots across ALL cell types, one panel per caste

import scanpy as sc
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------
# SETTINGS
# -----------------------
celltype_col = "paper_cell_type_annotation"
caste_col = "caste"

caste_order = ["worker", "soldier", "king", "queen"]
genes_to_plot = [g for g in gene_info.keys()]

# make labels like: Znev00002332 (PGRP)
gene_labels = {}
for g in genes_to_plot:
    desc = gene_info.get(g, g)
    short = desc.split("–")[0].strip() if "–" in desc else desc.strip()
    gene_labels[g] = f"{g} ({short})"

# use ALL annotated cell types
all_celltypes = (
    adata.obs[celltype_col]
    .astype(str)
    .replace("nan", pd.NA)
    .dropna()
    .unique()
    .tolist()
)

# stable order
celltype_order = sorted(all_celltypes)

# -----------------------
# One Scanpy dotplot per CASTE
# -----------------------
for caste in caste_order:
    adata_c = adata[adata.obs[caste_col] == caste].copy()

    genes_present = [g for g in genes_to_plot if g in adata_c.var_names]
    if len(genes_present) == 0:
        print(f"[skip] caste={caste}: none of the requested genes are in adata.var_names")
        continue

    cts_present = (
        adata_c.obs[celltype_col]
        .astype(str)
        .replace("nan", pd.NA)
        .dropna()
        .unique()
        .tolist()
    )

    ct_order_this = [ct for ct in celltype_order if ct in set(cts_present)]

    adata_c.obs[celltype_col] = pd.Categorical(
        adata_c.obs[celltype_col].astype(str),
        categories=ct_order_this,
        ordered=True,
    )

    print(f"\n=== {caste} === (n_cells={adata_c.n_obs}, n_celltypes={len(ct_order_this)})")

    dp = sc.pl.dotplot(
        adata_c,
        var_names=genes_present,
        groupby=celltype_col,
        categories_order=ct_order_this,
        color_map="viridis",
        dot_max=0.4,
        vmax=1.5,
        standard_scale=None,
        title=f"{caste}",
        show=False,
        swap_axes=True,
        figsize=(10, 2.8),
        return_fig=True,
    )

    ax_dict = dp.get_axes()

    if "mainplot_ax" in ax_dict:
        ax = ax_dict["mainplot_ax"]
        ax.set_yticklabels([gene_labels[g] for g in genes_present])

    dp.show()
```
